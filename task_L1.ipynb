import pandas as pd

```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset .csv")
df.head()
```

```python
df.columns
```

```python
df['Cuisines'] = df['Cuisines'].fillna('')
df['Cuisines'] = df['Cuisines'].str.split(',')
```

```python
df_exploded = df.explode('Cuisines')
```

```python
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.strip()
```

```python
cuisine_counts = df_exploded['Cuisines'].value_counts()

top_3_cuisines = cuisine_counts.head(3)

print(top_3_cuisines)
```

```python
total_restaurants = len(df)

top_3_percentages = (top_3_cuisines / total_restaurants) * 100

print(top_3_percentages)
```

```python
result = pd.DataFrame({
    'Cuisine': top_3_cuisines.index,
    'Number of Restaurants': top_3_cuisines.values,
    'Percentage (%)': (top_3_cuisines.values / total_restaurants) * 100
})

result
```
