```python
import pandas as pd

df = pd.read_csv("Dataset .csv")
df.head()
```

```python
df.columns
```

```python
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
```

```python
city_counts = df['City'].value_counts()
```

```python
top_city = city_counts.idxmax()
top_city_count = city_counts.max()
```

```python

print("Top City:", top_city)
print("Number of Restaurants:", top_city_count)

```

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

print("\nAverage Rating per City:\n", city_avg_rating)
```

```python
best_city = city_avg_rating.idxmax()
best_city_rating = city_avg_rating.max()

print("\nBest City by Rating:", best_city)
print("Average Rating:", best_city_rating)
```
