```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
df.head()
```

```python
df.columns
```

```python
price_counts = df['Price range'].value_counts().sort_index()

print(price_counts)
```

```python
total = len(df)

price_percentage = (price_counts / total) * 100

print(price_percentage)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x=price_counts.index, y=price_counts.values)

plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Price range'], bins=4, edgecolor='Red')

plt.title("Histogram of Price Range")
plt.xlabel("Price Range")
plt.ylabel("Frequency")

plt.show()
```

```python
result = pd.DataFrame({
    'Price Range': price_counts.index,
    'Number of Restaurants': price_counts.values,
    'Percentage (%)': price_percentage.values
})

result
```
